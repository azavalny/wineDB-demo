```python
import pandas as pd

vineyards = pd.read_csv('vineyard.csv')
vineyards.head()
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
import time

chrome_options = Options()
#chrome_options.add_argument("--headless")
chrome_options.add_argument("--disable-gpu")
chrome_options.add_argument("--no-sandbox")

driver = webdriver.Chrome(options=chrome_options)

# Load the CellarTracker search page
url = "https://www.cellartracker.com/list.asp?fInStock=0&iUserOverride=0&Table=List&szSearch=Nicosia+Etna+Vulk%E0+Bianco+Italy"
driver.get(url)

time.sleep(1)

# Get page source after JS rendering
html = driver.page_source
driver.quit()

# Parse HTML with BeautifulSoup
soup = BeautifulSoup(html, "html.parser")
table = soup.find("table", {"id": "main_table"})
wine_list = []

# Parse each table row
for row in table.find_all("tr")[1:]:
    cols = row.find_all("td")
    if len(cols) < 2:
        continue

    wine_type = cols[0].get_text(strip=True)

    name_block = cols[1]
    name = name_block.find("h3").get_text(strip=True)
    loc = name_block.find("span", class_="el loc").get_text(strip=True)
    var = name_block.find("span", class_="el var").get_text(strip=True)

    qty = cols[2].find("span", class_="el gty")
    quantity = qty.get_text(strip=True) if qty else ""

    score_span = cols[3].find("span", class_="el scr")
    score = score_span.get_text(strip=True) if score_span else ""

    wine_list.append({
        "type": wine_type,
        "name": name,
        "location": loc,
        "varietal": var,
        "quantity": quantity,
        "score": score
    })

# Print a few results
for wine in wine_list[:3]:
    print(wine)
```
